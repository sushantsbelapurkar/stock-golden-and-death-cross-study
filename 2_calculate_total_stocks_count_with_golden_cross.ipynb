```python
import pandas as pd
from os import listdir
from os.path import isfile, join
import numpy as np
```

```python
def is_golden(mva_50_l, mva_200_l):
    if len(mva_50_l) <= 0 or len(mva_200_l) <= 0:
        return False
    # pick golden cross
    status = 0
    cur_price_50 = mva_50_l[-1]
    for i in reversed(range(len(mva_50_l))):
        price_50 = mva_50_l[i]
        price_200 = mva_200_l[i]
        if status == 1:
            if price_200 > price_50:
                if cur_price_50 > (price_50 + price_200) / 2:
                    return True
                else:
                    return False
        else:
            if price_200 > price_50:
                status = -1
            else:
                status = 1
    return False
```

```python
mypath = 'full_history'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
i = 0
golden_stocks = []
for f in onlyfiles:
    i += 1
    if i % 500 == 0:
        print(i)
    # load in data
    path = 'full_history/' + f
    df_1 = pd.read_csv(path)
    df_1 = df_1.sort_values('date')
    # get moving average
    mva_50 = df_1['close'].rolling(50).mean()
    mva_200 = df_1['close'].rolling(200).mean()
    mva_50_l = list(mva_50)[199:]
    mva_200_l = list(mva_200)[199:]
    result = is_golden(mva_50_l, mva_200_l)
    if result == True:
        golden_stocks.append(f)
```
